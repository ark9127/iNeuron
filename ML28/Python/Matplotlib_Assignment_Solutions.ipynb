__Problem Statement 1:__

__Scipy:__
We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.

Task:

1. fitting it to the periodic function
2. plot the fit

Data

    Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
    Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

__Solution 1:__

```python
# Importing libraries and preparing data

import numpy as np
import matplotlib.pyplot as plt

temp_max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
temp_min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
# defining periodic function

from scipy import optimize

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, temp_min, [-40, 20, 0])
```

```python
# Plotting a fit

days = np.linspace(0, 12, num=365)
months = np.arange(12)

plt.figure()
plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'

titanic = pd.read_csv(url)

titanic.head()
```

__Solution 2.1:__

```python
titanic_sex = titanic['sex'].unique()
titanic_sex_val = titanic['sex'].value_counts()
titanic_sex, titanic_sex_val
```

```python
df = pd.DataFrame(titanic['sex'].value_counts(), index =  titanic['sex'].unique())
df = df.dropna()
df
```

```python
df.plot.pie(subplots = True, figsize = (5, 5), fontsize = 20, autopct = '%.2f%%', title = 'Male/Female Proportion')
plt.title
```
